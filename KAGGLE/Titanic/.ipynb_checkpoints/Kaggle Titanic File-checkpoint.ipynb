```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'C:\Users\anush\OneDrive\Desktop\NARESH IT\DATA SCIENCE PROJECTS\KAGGLE\Titanic\Titanic-Dataset.csv')
df.head(1)
```

```python
data = df[['Survived','Pclass','Fare']]
xtrain,xtest,ytrain,ytest = train_test_split(data.drop(columns='Survived'),
                                             data['Survived'],test_size=0.3,random_state=100)
LR = LogisticRegression()
LR_base = LR.fit(xtrain,ytrain)
base_pred = LR_base.predict(xtest)
accuracy_score(ytest,base_pred)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# lets use the LR base model to predict the test data from Kaggle.
test = pd.read_csv('test.csv')
test_data = test[['Pclass','Fare']]
test_data.isnull().sum()
test_data['Fare'].fillna(df['Fare'].mean(),inplace=True)
test_predictions = pd.DataFrame(LR_base.predict(test_data),columns=['Survived'])
```

```python
LR_base_model_predictions = pd.concat([test[['PassengerId']],test_predictions],axis=1)
```

```python
LR_base_model_predictions.to_csv('submission_1.csv')
```
