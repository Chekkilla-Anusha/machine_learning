```python
import os

base_path = "vest_helmet"

for root, dirs, files in os.walk(base_path):
    print(root)
```

```python
with open("vest_helmet/data.yaml", "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
results = model.train(
    data="vest_helmet/data.yaml",
    epochs=10,
    imgsz=640,
    batch=16,
    project="runs",
    name="ppe_project"
)
```

```python
metrics = model.val()

print("mAP50-95 :", metrics.box.map)
print("mAP50    :", metrics.box.map50)
print("Precision:", metrics.box.mp)
print("Recall   :", metrics.box.mr)
```

```python
results = model.predict(
    source="sample.jpg",
    save=True,
    conf=0.25
)
```
